# 매출데이터 전처리

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv('data/상권-추정매출_merged.csv', encoding='cp949')
```

```python
stores_code = pd.read_csv("data/상권-상권영역.csv", encoding="cp949")
```

```python
# 상권 구분 정리
sales_df = sales_df[sales_df['상권_구분_코드_명'].isin(['골목상권', '발달상권'])]
```

```python
# 2022 이전 상권 코드 일원화
# 2022년 이전/ 이후 상권 코드 1000000 씩 차이 남

sales_df['상권_코드'] = sales_df['상권_코드'].apply(lambda x: x + 1000000 if x < 3000000 else x)
```

```python
# 시점 컬럼, 문자열
sales_df['집계시점'] = sales_df['기준_년_코드'].astype(str) + '-Q' + sales_df['기준_분기_코드'].astype(str)
```

```python
# 모든 상권 코드 있는지 확인
sales_df[~sales_df['상권_코드'].isin(stores_code['상권_코드'])]
```

```python
# 상권 코드명 정리 (NaN 값 및 불필요한 표기 이상 제거)
sales_df['상권_코드_명'] = sales_df['상권_코드'].map(dict(zip(stores_code['상권_코드'], stores_code['상권_코드_명'])))
```

```python
sales_df.isna().any().any()
```

```python
# 점포 수 3개 미만은 마스킹 처리되어 있음 (NaN 값이 아님)
# 현재 시점에서는 그냥 두도록 함
sales_df[sales_df['점포수'] == '*'].head(3)
```

## 업종 재분류
100대 생활밀접업종을 연구에 용이한 형태로 재분류,  
분류 기준은 [서울연구원의 연구](https://www.si.re.kr/node/66029)를 참고하였음.

```python
reclassifier = pd.read_csv('data/업종재분류_기준표.csv', encoding='cp949')
reclassifier.head()
```

```python
# 중분류 오류 확인
# print 값 없으면 OK

unique_class_names = sales_df['서비스_업종_코드_명'].unique()

for name in reclassifier['중분류']:
    if name not in unique_class_names: 
        print(name)
```

```python
# 분류, 없는 값은 기타로 처리
reclassifier = dict(zip(reclassifier['중분류'], reclassifier['재분류']))
sales_df['업종_재분류'] = sales_df['서비스_업종_코드_명'].map(reclassifier).fillna('기타')
```

```python
sales_df.head(3)
```

```python
grouping_scheme = ["집계시점", "상권_구분_코드", "상권_구분_코드_명", "상권_코드", "상권_코드_명", "업종_재분류"]
```

```python
class_grouped = sales_df.groupby(grouping_scheme).sum().reset_index()
class_grouped.drop(['기준_년_코드', '기준_분기_코드'], axis=1, inplace=True)
class_grouped
```

```python
# 저장
class_grouped.to_csv('data/상권-추정매출_업종재분류.csv', index=False, encoding='cp949')
```

##  분석용 데이터 프레임

```python
# 현재 상권영역 기반으로 데이터 작업
district_df = pd.read_csv('data/상권-상권영역.csv', encoding='cp949')
```

```python
district_df.drop('기준_년월_코드', axis=1, inplace=True)
```

```python
# 상권 구분 정리
district_df = district_df[district_df['상권_구분_코드_명'].isin(['골목상권', '발달상권'])]
```

```python
# 상권 코드로 인덱스 설정 (빨라집니다)
district_df.set_index('상권_코드', inplace=True)
```

```python
district_df
```

### 시군구명 추가

```python
seoul_admin_info = pd.read_csv('data/서울시 건축물대장 법정동 코드정보.csv', encoding='cp949')
```

```python
gu_dict = dict(zip(seoul_admin_info['시군구코드'], seoul_admin_info['시군구명']))
dong_dict = dict(zip(seoul_admin_info['시군구코드'] * 1000 + seoul_admin_info['행정동코드'], seoul_admin_info['행정동명']))
```

```python
district_df['시군구_명'] = district_df['시군구_코드'].map(gu_dict)
district_df['행정동_명'] = district_df['행정동_코드'].map(dong_dict)
```

```python
district_df.head(2)
```

```python
SECTORS = class_grouped['업종_재분류'].unique()
SECTORS
```

```python
merge_mask = class_grouped.groupby(['상권_코드', '업종_재분류']).size().reset_index()
merge_mask = merge_mask[['상권_코드', '업종_재분류']]
merge_mask
```

```python
merged_df = pd.merge(district_df, merge_mask, on='상권_코드', how='left')
merged_df.head(3)
```

```python
timestamps = [f"{year}-Q{quarter}" for year in range(2017, 2023) for quarter in range(1, 5)] + ['2023-Q1']
```

```python
# 분기당 매출 금액 컬럼 생성
merged_df.set_index(['상권_코드', '업종_재분류'], inplace=True)
class_grouped.set_index(['상권_코드', '업종_재분류'], inplace=True)
for timestamp in timestamps:
    merged_df[timestamp] = merged_df.index.map(class_grouped[class_grouped['집계시점'] == timestamp]['분기당_매출_금액'])
```
